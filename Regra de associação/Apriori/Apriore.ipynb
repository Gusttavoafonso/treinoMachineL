# 🧠 Apriori Algorithm

O **Apriori** é um algoritmo clássico de mineração de dados usado para extrair **regras de associação** entre itens em grandes bases de dados transacionais, como vendas de supermercados.

Seu objetivo é identificar padrões do tipo:

> **Se** um cliente compra {Leite, Pão} → **então** ele também compra {Manteiga}

---

## 📌 Como Funciona

O Apriori segue três etapas principais:

1. **Gerar conjuntos de itens frequentes** (frequent itemsets)
   - Um conjunto de itens é frequente se seu **suporte** for maior que um valor mínimo definido pelo usuário.

2. **Explorar a propriedade Apriori**:
   - Se um itemset é infrequente, todos os seus superconjuntos também serão infrequentes.
   - Isso reduz drasticamente o espaço de busca.

3. **Gerar regras de associação** a partir dos conjuntos frequentes
   - Com base em métricas como **confiança** e **lift**.

---

## 📐 Métricas Utilizadas


- **Suporte** (Support):

$$
\text{Support}(A) = \frac{\text{Número de transações contendo } A}{\text{Número total de transações}}
$$

- **Confiança** (Confidence):

$$
\text{Confidence}(A \rightarrow B) = \frac{\text{Support}(A \cup B)}{\text{Support}(A)}
$$

- **Lift**:

$$
\text{Lift}(A \rightarrow B) = \frac{\text{Confidence}(A \rightarrow B)}{\text{Support}(B)}
$$

---

✅ Vantagens
Simples e fácil de implementar

Reduz espaço de busca com base em regras matemáticas

Boa performance em conjuntos de dados pequenos a médios

⚠️ Desvantagens
Pode ser lento em bases de dados muito grandes

Gera muitos candidatos (combinações de itens) → custo computacional alto

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
