# Polinominal regressao

## importando bibliotecas


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## importando dataset

```python
ds = pd.read_csv("Position_Salaries.csv")
X = ds.iloc[ : , 1 : -1].values
y = ds.iloc[:, -1].values
```

## treinando a regressao linear

```python
# treinar regressao linear para mostrar que não se encaixa bem

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)

from sklearn.linear_model import LinearRegression
Lregressor = LinearRegression()
Lregressor.fit(X, y)
```

## treinando regressao polinominal

```python
from sklearn.preprocessing import PolynomialFeatures
PoRegressor = PolynomialFeatures(degree= 4)
X_poli = PoRegressor.fit_transform(X)
lin_2 = LinearRegression()
lin_2.fit(X_poli, y)
```

## visualizando resultados

```python
#regressao linear
plt.scatter(X, y, color = 'pink')
plt.plot(X,Lregressor.predict(X), color = 'red')
plt.title('Verdadeiro ou falso (Regressao linear)')
plt.xlabel('posição de nivel')
plt.ylabel("salario")
plt.show()
```

```python
#regressao Polinominal
plt.scatter(X, y, color = 'pink')
plt.plot(X,lin_2.predict(X_poli), color = 'red')
plt.title('Verdadeiro ou falso (Regressao linear)')
plt.xlabel('posição de nivel')
plt.ylabel("salario")
plt.show()
```

```python
# deixando as curvas mais suavez
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_2.predict(PoRegressor.fit_transform(X_grid)), color = 'blue')
plt.title('verdadeiro ou falso (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
